### some problems may encounter (edge cases)
- `query` is the first element in `cards`
- `query` is the last element in `cards`
- the list `cards` contains only one element
- the list `cards` does not contain the query
- the list `cards` is empty
- the list `cards` contains duplicate elements
- the number `query` occurs multiple times in `cards`

```python
tests = []
# normal case
tests.append({
    'input': {
        'cards': [13, 11, 10, 7, 4, 3, 1, 0],
        'query': 7
    },
    'output': 3
})
# normal case
tests.append({
    'input': {
        'cards': [13, 11, 10, 7, 4, 3, 1, 0],
        'query': 1
    },
    'output': 6
})
# query is the first card
tests.append({
    'input': {
        'cards': [4, 2, 1, -1],
        'query': 4
    },
    'output': 0
})
# query is the last card
tests.append({
    'input': {
        'cards': [3, -1, -9, -127],
        'query': -127
    },
    'output': 3
})
# cards has only one element
tests.append({
    'input': {
        'cards': [6],
        'query': 6
    },
    'output': 0
})
# query not in cards
tests.append({
    'input': {
        'cards': [9, 7, 5, 2, -9],
        'query': 4
    },
    'output': -1
})
# cards has no element
tests.append({
    'input': {
        'cards': [],
        'query': 7
    },
    'output': -1
})
# cards has duplicate elements
tests.append({
    'input': {
        'cards': [1, 1, 1],
        'query': 1
    },
    'output': 0
})
```

> the simplest or most obvious solution to a problem is called a brute-force solution. It is often used as a baseline solution to check the correctness of other solutions.

```python
# linear search
def locate_cards_v1(cards, query):
    position = 0
    for card in cards:
        if card == query:
            return position
        position += 1
    return -1

# linear search
def locate_cards_v2(cards, query):
    position = 0
    while position < len(cards):
        if cards[position] == query:
            return position
        position += 1
    return -1
```

## Complexity and Big O notation
> In the Big O, we drop the constant factor and consider the highest order term. For example, if the time complexity of an algorithm is 5n^2 + 3n + 2, then the time complexity of the algorithm is O(n^2).

```python
def locate_cards_v3(cards, query):
    low, high = 0, len(cards) - 1
    while True:
        if low > high:
            return -1
        mid = (low + high) // 2
        if cards[mid] == query:
            if mid == 0 or cards[mid - 1] != query:
                return mid
            else:
                high = mid - 1
        elif cards[mid] < query:
            high = mid -1
        else:
            low = mid + 1
```
